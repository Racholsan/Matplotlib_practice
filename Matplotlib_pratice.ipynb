```python
from matplotlib import pylab
print(pylab.__version__)
```

```python
""" Use Numpy to generate Random data
"""

import numpy as np
x = np.linspace(0, 10, 25)
y = x * x +2
print(x)
print(y)
```

```python
"""It only takes 1 command to draw
"""
pylab.plot(x, y , 'r') # r stands for red
```

```python
"""Drawing a subgraph
"""
pylab.subplot(1, 2, 1) # The contents of the bracket represents (rows, columns, indexes)
pylab.plot(x, y, 'r--')
pylab.subplot(1, 2, 2)
pylab.plot(y, x, 'g*-')
```

```python
# (1, 2, 1) and (1, 2, 2) are significant if want tiled subplots, without which both the plots will be in same plot

# for example

pylab.subplot(1, 2, 1) # The contents of the bracket represents (rows, columns, indexes)
pylab.plot(x, y, 'r--')
pylab.subplot(1, 2, 1)
pylab.plot(y, x, 'g*-')
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()

axis = fig.add_axes([0.5, 0.1, 0.8, 0.8]) # Control the left, right, width, height of canvas (from 0 to 1)

axis.plot(x, y, 'r')
```

```python
# again we can draw subplots

fig, axes = plt.subplots(nrows = 1, ncols = 2) # Subplot is of 1 row, 2 columns

for ax in axes:
    ax.plot(x, y, 'r')
```

```python
# We can also draw a picture, or graph inside another graph

fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # big axes
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # small canvas

axes1.plot(x, y, 'r')
axes2.plot(y, x, 'g')
```

```python
fig = plt.figure(figsize = (16, 9), dpi = 300) # New graphic object

fig.add_subplot()

plt.plot(x, y, 'r')
```

```python
fig, axes = plt.subplots(figsize=(16,9), dpi= 300)
axes.set_title("TITLE", fontsize = 20)
axes.set_xlabel("X", fontsize = 18)
axes.set_ylabel("Y", fontsize = 18)

axes.plot(x, x**2)
axes.plot(x, x**3)

axes.legend(["y = x**2", "y = x**3"], loc = 2, fontsize  = 18)

```

```python
# In Matplotlib, you can set other properties such as line color, transparency, and more

fig, axes = plt.subplots(dpi = 150)
axes.plot(x, x+1, color = 'red', alpha = 0.5)
axes.plot(x, x+2, color = 'blue')
axes.plot(x, x+3, color = 'green')
```

# Other 2D graphics


```python
n = np.array([0, 1, 2, 3, 4, 5])

fig, axes = plt.subplots(1, 4, figsize = (16, 5))

axes[0].set_title("scatter")
axes[0].scatter(x, x+0.25*np.random.randn(len(x)))

axes[1].set_title("step")
axes[1].step(n, n**2, lw = 2)

axes[2].set_title("bar")
axes[2].bar(n, n**2, align = 'center', width = 0.5, alpha  = 0.5)

axes[3].set_title("fill_between")
axes[3].fill_between(x, x**2, x**3, color = 'green', alpha = 0.5)
```

```python
# Draw a radar chart

fig = plt.figure(figsize = (6,6))
ax = fig.add_axes([0.0, 0.0, 1, 1], polar = True)
t = np.linspace(0, 2*np.pi, 100)
ax.plot(t, t, color = 'blue', lw = 3)
```

```python
""" Draw a histogram
"""

n = np.random.randn(100000)
fig, axes = plt.subplots(1, 2, figsize = (12, 4))
axes[0].set_title("Default histogram")
axes[0].hist(n)

axes[1].set_title("Cumulative Detailed Histogram")
axes[1].hist(n, cumulative = True, bins = 50)
```

```python
# Draw contour image

import matplotlib
import numpy as np
import matplotlib.cm as cm
import matplotlib.pyplot as plt

delta = 0.025
x = np.arange(-3.0, 3.0, delta)
y = np.arange(-2.0, 2.0, delta)
X, Y = np.meshgrid(x, y)
Z1 = np.exp(-X**2 - Y**2)
Z2 = np.exp(-(X-1)**2 - (Y-1)**2)
Z = (Z1 - Z2)*2

print(X)
print(Y)
```

```python
fig, ax = plt.subplots()
CS = ax.contour(X, Y, Z)
ax.clabel(CS, inline = 1, fontsize = 10)
ax.set_title("contour map")
```

```python
# Draw 3D surface image

from mpl_toolkits.mplot3d.axes3d import Axes3D

fig  = plt.figure(figsize = (14, 6))

# Specify the 3D graphics to draw, with projection = '3d'
ax = fig.add_subplot(1, 2, 1, projection = '3d')
ax.plot_surface(X, Y, Z, rstride =4, cstride =4, lw = 0)
```

# Write a python program to create a pie chart of the popularity of programming languages

```python
import matplotlib.pyplot as plt
# Data to plot
languages  = 'Python', 'Java', 'PHP', 'Javascript', 'C#', 'C++'
Popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
colors = ['green', 'red', 'yellow', 'orange', 'blue', 'black']
```

```python
# Explode 1st slice
explode = (0.1, 0, 0, 0, 0, 0)

plt.pie(Popularity)
```

```python
# Explode 1st slice
explode = (0.1, 0, 0, 0, 0, 0)

plt.pie(Popularity, explode = explode, labels = languages, colors = colors,
        autopct = '%1.1f%%', shadow = True, startangle = 140)

```
